```python
import pandas as pd
```

```python
results = pd.read_csv('results/1D_synth_reweight_NS_10000_NE_20_RUNS_10.csv')
results.drop(columns=['Unnamed: 0', 'AU-ROC'], inplace=True)
        
```

```python

def plot_exp1_results(results, stat):
    if stat == 'mean':
        res = results.groupby(['alpha', 'beta', 'model']).mean().reset_index()
    else:
        res = results.groupby(['alpha', 'beta', 'model']).sem().reset_index()
    res = res.pivot(index="model", columns=['alpha', 'beta'])
    res = res.reindex(["OBS", "COM", "RW", "COM-SL", "RW-SL", "Proxy Oracle", "Target Oracle"])
    res = res * 100
    
    return res
```

```python
res = plot_exp1_results(results, 'mean')
res
# res.style.format('{:.1f}')
```

```python
res = plot_exp1_results(results, 'sem')
res.style.format('{:.1f}')
```
